```python
import numpy as np
import pandas as pd
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, LabelEncoder


newest_year = 2006

df = pd.read_csv('spreadspoke_scores.csv')
#encoding team names
encoder = OneHotEncoder(sparse_output=False)
encoded = encoder.fit_transform(df[['team_home', 'team_away']])
new_df = pd.DataFrame(encoded, columns=encoder.get_feature_names_out(['team_home', 'team_away']))

#dropping old data and games that haven't been played yet
df = pd.concat([df, new_df], axis=1).drop(df[df.score_home.isnull()].index).drop(df[df.schedule_season < newest_year].index)

#encoding true/false values
encoder = LabelEncoder()
df['stadium_neutral'] = encoder.fit_transform(df['stadium_neutral'])

#dropping columns that are not needed
# df['schedule_date'] = pd.to_datetime(df['schedule_date'], errors='coerce')
# df['year'] = df['schedule_date'].dt.year
df = df.drop(columns=['schedule_date', 'schedule_playoff', 'over_under_line', 'team_favorite_id', 'spread_favorite', 'stadium', 'weather_detail'])

#fixing missing values
df['weather_temperature'] = df['weather_temperature'].fillna(df['weather_temperature'].mean())
df['weather_wind_mph'] = df['weather_wind_mph'].fillna(df['weather_wind_mph'].mean())
df['weather_humidity'] = df['weather_humidity'].fillna(df['weather_humidity'].mean())

#Making sure week is a number
df['schedule_week'] = df['schedule_week'].replace(['Wildcard','Division', 'Conference', 'SuperBowl'], [18, 19, 20, 21])
df = df.reset_index(drop=True)

player_data = pd.read_csv('NFL_AllPlayers_Dataset.csv')

df = df.join(player_data.set_index(['season', 'week']), on=['schedule_season', 'schedule_week'], how='left')

df.to_csv('output.csv', index=False)
player_data

#.drop(columns=['team_home', 'team_away'])
```
